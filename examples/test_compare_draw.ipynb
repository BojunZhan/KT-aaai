```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.figure(figsize=(10,6), dpi=100)
auc_89M = [0.7787, 0.7856, 0.7844]
auc_44M = [0.7843, 0.7849, 0.7796]
auc_22M = [0.7786, 0.7814, 0.7777]
auc_221M = [0.7876, 0.7993, 0.7868]
# 按照seq_len200，512，1024 设置画图的x轴
seq_200 = [auc_22M[0], auc_44M[0], auc_89M[0], auc_221M[0]]
seq_512 = [auc_22M[1], auc_44M[1], auc_89M[1], auc_221M[1]]
seq_1024 = [auc_22M[2], auc_44M[2], auc_89M[2], auc_221M[2]]
labels = ["param_22M", "param_44M", "param_89M", "param_221M"]
# 画图,不同的auc用不同的颜色区分，并且设置图例
plt.bar(np.arange(4), height=seq_200, width=0.2, label="seq200", edgecolor="white", color="red", tick_label=labels)
plt.bar(np.arange(4)+0.2, height=seq_512, width=0.2, label="seq512", edgecolor="white", color="grey")
plt.bar(np.arange(4)+0.4, height=seq_1024, width= 0.2, label="seq1024", edgecolor="white", color="skyblue")
# 设置x轴的刻度
plt.xticks(np.arange(4)+0.13,labels ,fontsize = 12)
# 设置x轴的名称
plt.xlabel('params_num')
# 设置y轴的名称
plt.ylabel('test_auc')
plt.ylim(0.65, 0.8)
plt.yticks(np.arange(0.65,0.8,0.005))
# 设置图例
plt.legend()
plt.title('AS2009')
# 显示图像
plt.show()
```
