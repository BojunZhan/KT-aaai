```python
import re
import csv

# 打开csv文件
with open('88M-seq1024-train_loss.csv', 'r') as file:
    # 创建一个csv读取器
    reader = csv.reader(file)
    
    # 创建一个空列表来保存提取的数字
    train_losses = []
    next(reader)
    # 遍历每一行数据
    for row in reader:
        # 使用正则表达式提取所有数字部分
        nums = re.findall(r'\d+\.?\d*', row[0])
        
        # 将找到的数字转换为浮点数，并添加到列表中（删除空格）
        train_losses.append([float(num.replace(" ", "")) for num in nums])
    
    # 打印提取的数字
    print(train_losses)
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.plot(train_losses)
plt.xlabel('epoch')
plt.ylabel('loss')
# y轴划分到小数点
plt.title('89M_512_train_loss')
max_loss = np.max(train_losses)
min_loss = np.min(train_losses)
min_loss = round(min_loss, 3)
epoch_len = len(train_losses)
plt.yticks(np.arange(min_loss, max_loss, 0.001))
plt.xticks(np.arange(0, epoch_len, 2))
plt.show()
```
