```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data_as = pd.read_csv('pred_result/dkt_as09.csv')
as_test_auc = data_as['windowauclate_mean']
data_nips = pd.read_csv('pred_result/dkt_nips.csv')
nips_test_auc = data_nips['windowauclate_mean']
data_al = pd.read_csv('pred_result/dkt_al05.csv')
al_test_auc = data_al['windowauclate_mean']
data_bd = pd.read_csv('pred_result/dkt_bd06.csv')
bd_test_auc = data_bd['windowauclate_mean']
data_xes = pd.read_csv('pred_result/dkt_peiyou.csv')
xes_test_auc = data_xes['windowauclate_mean']
# 创建一个列表，其中包含所有的AUC数据
colors = ['3468C0', '86A7FC', 'FFDD95', 'FF9843', 'D9EDBF']
data = [as_test_auc, nips_test_auc, al_test_auc, bd_test_auc, xes_test_auc]
unikt_data = [0.7706, 0.8004, 0.8336, 0.8018, 0.8167] 
# 创建一个列表，其中包含所有的数据集标签
labels = ['AS09','NIPS', "AL05", "BD06", "XES3G5M"]
# 创建一个figure和axes
fig, ax = plt.subplots()
# 对于每个数据集
for i in range(len(data)):
    # 创建一个散点图
    ax.scatter(data[i], [labels[i]]*len(data[i]),color='blue', marker='o', alpha=0.5)
    ax.scatter(unikt_data[i], labels[i], color='red', marker='x')
# 设置标题和轴标签
ax.set_title('AUC Distribution for Each Dataset')
ax.set_xlabel('AUC')
ax.set_ylabel('Dataset')

# 显示图形
plt.show()

```
